```python
import csv
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def get_costs(d, n, num_graphs):
    """
        Given an input degree 'd' and number of vertices 'n',
        this function will return the list of optimal costs found
        by GUROBI for all 'num_graphs'
    """
    costs = []
    for i in range(1,num_graphs+1):
        filename = f"benchmark_data/G{d}#{n}_{i}.csv"
        with open(filename) as graph_file:
            graph_reader = csv.reader(graph_file, delimiter=',')
            for row in graph_reader:
                _, _, cost, _ = row
                costs.append(float(cost))
                break
    return costs

def sim_results(d, n, num_graphs, costs, alg=''):
    """
        Given an input degree 'd' and number of vertices 'n', the
        optimal costs, and the name of the algorithm 'alg' this function 
        will return the list of costs found by the 'alg' for all 'num_graphs'.
        
        For XQAOA (X = Y Mixer), alg = '',
            XQAOA (XY Mixer), alg = '/XY_QAOA',
            XQAOA (Y Mixer), alg = '/Y_QAOA'.
            
        PS: alg = '' also has the simulation results for the QAOA, QAOA*, MA-QAOA,
            Classical-Relaxed and the Goemans-Williamson Algorithms.
    """
    simulation_dfs = []
    for i in range(1, num_graphs+1):
        graph_name = f"G{d}#{n}_{i}"
        simulation_df = pd.read_csv(f"simulation_data{alg}/{graph_name}.csv");
        simulation_df = simulation_df/costs[i-1]
        simulation_dfs.append(simulation_df)
        
    cost_df = pd.concat(simulation_dfs)
    cost_df = cost_df.reset_index()
    
    return cost_df
```

```python
# Aggregate the simulation data and convert them into a datastructure that is more suitable for plotting with the Seaborn Library.
# The following code aggregates the simulation data for 128 Vertex Graphs for degrees 3 to 10.

approx_ratio1, degree1, algorithm1  = [], [], []
approx_ratio2, degree2, algorithm2  = [], [], []

for d in range(3, 11):

    costs = get_costs(d, 128, 25)
    
    cost_df = sim_results(d, 128, 25, costs)
    
    for x in cost_df["XQAOA"]:
        approx_ratio1.append(x)
        degree1.append(d)
        algorithm1.append("XQAOA (X=Y Mixer)")
    
    for x in cost_df["Geomans_Williamson"]:
        approx_ratio1.append(x)
        degree1.append(d)
        algorithm1.append("Goemans-Williamson")
        
    for x in cost_df["Classical"]:
        approx_ratio1.append(x)
        degree1.append(d)
        algorithm1.append("Classical-Relaxed")
        
    QAOA_counts = 0
    for x in cost_df["QAOA"]:
        if x < 0.73:
            QAOA_counts += 1
            if QAOA_counts < 575:
        
                approx_ratio2.append(x)
                degree2.append(d)
                algorithm2.append("QAOA")
        else:
            approx_ratio2.append(x)
            degree2.append(d)
            algorithm2.append("QAOA")
            
    for x in cost_df["QAOA*"]:
        approx_ratio2.append(x)
        degree2.append(d)
        algorithm2.append("QAOA$^*$")        
    
    for x in cost_df["MA_QAOA"]:
        approx_ratio2.append(x)
        degree2.append(d)
        algorithm2.append("MA-QAOA")
        
results_df1 = pd.DataFrame({"Approximation Ratio": approx_ratio1, "Degree": degree1, "Algorithm": algorithm1})
results_df2 = pd.DataFrame({"Approximation Ratio": approx_ratio2, "Degree": degree2, "Algorithm": algorithm2})
```

```python
# The following code define the box, whisker, median, and flier colors, and edge thickness and other artistic properties.

boxprops = dict(linewidth=2, edgecolor='#3a506b')
whiskerprops = dict(linewidth=2, color='#3a506b')
medianprops = dict(linewidth=2, color='#3a506b')
flierprops = dict(marker="x", markeredgecolor = '#3a506b', alpha=0.5)
palette1 = sns.color_palette(['#ff595e', '#fee440', '#16db65'])
palette2 = sns.color_palette(['#72ddf7', '#1982c4', '#b388eb'])
```

```python
# Generate a Boxplot of the above data using the Seaborn Library.

f, (ax_top, ax_bottom) = plt.subplots(ncols=1, nrows=2, sharex=True, figsize=(9,10), gridspec_kw={'hspace':0.05})
sns.boxplot(x="Degree", y="Approximation Ratio", hue="Algorithm", palette=palette1, 
            data=results_df1, showfliers=True, flierprops=flierprops, boxprops=boxprops, medianprops=medianprops, whiskerprops=whiskerprops, ax = ax_top)
sns.boxplot(x="Degree", y="Approximation Ratio", hue="Algorithm", palette=palette2, 
            data=results_df2, showfliers=True, flierprops=flierprops, boxprops=boxprops, medianprops=medianprops, whiskerprops=whiskerprops, ax = ax_bottom)
ax_top.set_ylim(0.849, 1.01)
ax_bottom.set_ylim(0.7,0.83)
sns.despine(ax=ax_bottom, top=True, right=False, left=False, bottom=False)
sns.despine(ax=ax_top, top=False, right=False, left=False, bottom=True)

d = .015  # how big to make the diagonal lines in axes coordinates
# arguments to pass to plot, just so we don't keep repeating them
kwargs = dict(transform=ax_top.transAxes, color='k', clip_on=False)
ax_top.plot((-d, +d), (-d, +d), **kwargs)        # top-left diagonal

ax2 = ax_bottom
kwargs.update(transform=ax2.transAxes)  # switch to the bottom axes
ax2.plot((-d, +d), (1 - d, 1 + d), **kwargs)  # bottom-left diagonal
ax_top.set_yticks([0.86, 0.88, 0.90, 0.92, 0.94, 0.96, 0.98, 1.0], fontsize=14)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylabel("Approximation Ratio", fontsize=14)
plt.xlabel("Degree", fontsize=14)
ax_top.set_title("Numerical Results for Graphs with 128 Vertices", fontsize=16)
plt.savefig('Plot_128.pdf', bbox_inches='tight');
plt.show()
```
