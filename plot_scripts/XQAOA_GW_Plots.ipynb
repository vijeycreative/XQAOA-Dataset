```python
import csv
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
def best_cuts(d, n=128):
    xqaoa = []
    gw = []
    
    costs = []
    for i in range(1,26):
        filename = f"benchmark_data/G{d}#{n}_{i}.csv"
        with open(filename) as graph_file:
            graph_reader = csv.reader(graph_file, delimiter=',')
            for row in graph_reader:
                _, _, cost, _ = row
                costs.append(float(cost))
                break
                
    for i in range(1, 26):
        graph_name = f"G{d}#{n}_{i}"
        simulation_df = pd.read_csv(f"simulation_data/{graph_name}.csv");
        simulation_df = simulation_df/costs[i-1]
        xqaoa.append(simulation_df["XQAOA"].max())
        gw.append(simulation_df["Geomans_Williamson"].max())
        
    return xqaoa, gw
```

```python
cost_df_XQAOA3, cost_df_GW3 = best_cuts(3)
cost_df_XQAOA4, cost_df_GW4 = best_cuts(4)
cost_df_XQAOA5, cost_df_GW5 = best_cuts(5)
cost_df_XQAOA6, cost_df_GW6 = best_cuts(6)
cost_df_XQAOA7, cost_df_GW7 = best_cuts(7)
cost_df_XQAOA8, cost_df_GW8 = best_cuts(8)
cost_df_XQAOA9, cost_df_GW9 = best_cuts(9)
cost_df_XQAOA10, cost_df_GW10 = best_cuts(10)
```

```python
plt.figure(figsize=(4,4))
plt.plot(np.linspace(0.969, 1.1, 50), np.linspace(0.969, 1.1, 50), c = "grey");
plt.scatter(cost_df_GW3, cost_df_XQAOA3, label = "$D = 3$", marker="x", c="#384abfff")
plt.scatter(cost_df_GW4, cost_df_XQAOA4, label = "$D = 4$", marker="x", c="#6687eeff")
plt.scatter(cost_df_GW5, cost_df_XQAOA5, label = "$D = 5$", marker="x", c="#99bafdff")
plt.scatter(cost_df_GW6, cost_df_XQAOA6, label = "$D = 6$", marker="x", c="#c7d7eeff")
plt.scatter(cost_df_GW7, cost_df_XQAOA7, label = "$D = 7$", marker="x", c="#eecfbfff")
plt.scatter(cost_df_GW8, cost_df_XQAOA8, label = "$D = 8$", marker="x", c="#f5a687ff")
plt.scatter(cost_df_GW9, cost_df_XQAOA9, label = "$D = 9$", marker="x", c="#e16851ff")
plt.scatter(cost_df_GW10, cost_df_XQAOA10, label = "$D = 10$", marker="x", c="#b30226ff")

plt.xticks([0.97, 0.98, 0.99, 1.0], fontsize=16)
plt.yticks([0.97, 0.98, 0.99, 1.0], fontsize=16)
plt.ylim(0.969, 1.001)
plt.xlim(0.969, 1.001)

#plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.05), ncol=8, fancybox=True, shadow=True)
plt.ylabel("XQAOA", fontsize=16)
plt.xlabel("GW", fontsize=16)
plt.savefig("Plot_128_Best.pdf",bbox_inches='tight')
```
